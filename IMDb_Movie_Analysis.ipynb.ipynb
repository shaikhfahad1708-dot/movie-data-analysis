````python
# 1) Split multi-genre rows into separate rows
g = (
    df.assign(Genre=df['Genre'].str.split(','))   # turn "A,B" into ["A","B"]
      .explode('Genre')                           # 1 row per genre
      .assign(Genre=lambda d: d['Genre'].str.strip())  # remove spaces
)

# 2) Group by genre and average only numeric columns we care about
genre_stats = (
    g.groupby('Genre')[['Rating','Metascore','Revenue (Millions)','Votes','Runtime (Minutes)']]
      .mean()
      .round(2)
      .sort_values('Rating', ascending=False)
)

genre_stats.head(10)  # see top 10 by average Rating
````
